## Linear Binary Classifier를 디자인하여 Logical OR 연산 출력

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# AND data

X_AND = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_AND = np.array([[0], [0], [0], [1]])
```

```python
# OR data

X_OR = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_OR = np.array([[0], [1], [1], [1]])
```

```python
# bias 추가

X_AND_ = np.c_[np.ones(X_AND.shape[0]), X_AND]
X_AND_
```

```python
# 초기 theta값 설정

theta = np.random.randn(3, 1)
```

```python
eta = 0.01
iterations = 1000
m = len(X_AND_)
```

```python
for iteration in range(iterations):
    gradients = 2/m * X_AND_.T.dot(X_AND_.dot(theta) - Y_AND)
    theta -= eta * gradients
```

```python
def linear_binary_classifier(X, theta):
    z = X.dot(theta)
    return (z >= 0).astype(int)
```

```python
Y_pred_AND = linear_binary_classifier(X_AND_, theta)
```

```python
# AND plot



plt.subplot(1, 2, 1)
plt.xlim(-0.2, 1.2)  
plt.ylim(-0.2, 1.2)
plt.plot(X_AND[Y_pred_AND[:, 0] == 0][:, 0], X_AND[Y_pred_AND[:, 0] == 0][:, 1], 'bs', markersize=10, label='AND False (Predicted)')
plt.plot(X_AND[Y_pred_AND[:, 0] == 1][:, 0], X_AND[Y_pred_AND[:, 0] == 1][:, 1], 'ro', markersize=10, label='AND True (Predicted)')
plt.title('Predicted AND Operation Results')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.legend()

# OR plot 

plt.subplot(1, 2, 2)
plt.xlim(-0.2, 1.2)  
plt.ylim(-0.2, 1.2)
plt.plot(X_OR[Y_OR[:, 0] == 0][:, 0], X_OR[Y_OR[:, 0] == 0][:, 1], 'bs', markersize=10, label='OR False')
plt.plot(X_OR[Y_OR[:, 0] == 1][:, 0], X_OR[Y_OR[:, 0] == 1][:, 1], 'ro', markersize=10, label='OR True')
plt.title('OR Operation Results')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
